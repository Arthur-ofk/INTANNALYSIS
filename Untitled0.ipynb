```python
import time, random, numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import DataLoader, Subset
import torchvision, torchvision.transforms as T
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_distances
from sklearn.metrics import precision_recall_fscore_support
from scipy.stats import entropy as shannon_entropy


DEVICE = "cuda"
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
# гіперпарметри AL з умови
INIT_LABELED = 200      # 5% старт
BATCH_AL = 100          # +100 за ітерацію
AL_STEPS = 10           # 10 ітерацій
ANNO_COST_PER_SAMPLE = 0.10
BATCH_SIZE = 128
EPOCHS = 2              # невеликі для швидкості
LR = 1e-3
TARGET_ACC = 0.90       # для Samples-to-Threshold

# CIFAR-10 + нормалізація
tr = T.Compose([T.ToTensor(),
                T.Normalize((0.4914,0.4822,0.4465),(0.2470,0.2435,0.2616))])

train_full = torchvision.datasets.CIFAR10(root="/content", train=True,  download=True, transform=tr)

# спліт 60/30/10 на основі train_full
idx = np.arange(len(train_full)); np.random.shuffle(idx)
n = len(idx); n_train = int(0.6*n); n_pool = int(0.3*n)
train_idx = idx[:n_train]
pool_idx  = idx[n_train:n_train+n_pool]
test_idx  = idx[n_train+n_pool:]       # 10%

train_ds = Subset(train_full, train_idx.tolist())
pool_ds  = Subset(train_full, pool_idx.tolist())
test_ds  = Subset(train_full, test_idx.tolist())
```

```python
#базова модель
class SmallCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.c1=nn.Conv2d(3,32,3,1,1)
        self.c2=nn.Conv2d(32,64,3,1,1)
        self.c3=nn.Conv2d(64,128,3,1,1)
        self.p=nn.MaxPool2d(2,2)
        self.fc1=nn.Linear(128*4*4,256)
        self.fc2=nn.Linear(256,10)
    def forward(self,x,return_feat=False):
        x=self.p(F.relu(self.c1(x)))
        x=self.p(F.relu(self.c2(x)))
        x=self.p(F.relu(self.c3(x)))
        x=x.view(x.size(0),-1)
        feat=F.relu(self.fc1(x))
        out=self.fc2(feat)
        return (out,feat) if return_feat else out

def create_model(seed=SEED):
    torch.manual_seed(seed)
    return SmallCNN().to(DEVICE)

def train_epoch(m,ldr,opt):
    # 1 епоха
    m.train()
    for x,y in ldr:
        x,y=x.to(DEVICE),y.to(DEVICE)
        opt.zero_grad()
        out=m(x)
        loss=F.cross_entropy(out,y)
        loss.backward()
        opt.step()

@torch.no_grad()
def evaluate(m,ldr):
    # acc + ймовірності + macro P/R/F1
    m.eval(); corr=0; P=[]; Y=[]
    for x,y in ldr:
        x=x.to(DEVICE); y=y.to(DEVICE)
        p=F.softmax(m(x),1)
        P.append(p.cpu().numpy()); Y.append(y.cpu().numpy())
        corr += (p.argmax(1)==y).sum().item()
    P=np.concatenate(P,0); Y=np.concatenate(Y,0)
    acc=corr/len(ldr.dataset)
    yhat=P.argmax(1)
    prec,rec,f1,_=precision_recall_fscore_support(Y,yhat,average="macro",zero_division=0)
    return acc,P,{"precision":prec,"recall":rec,"f1":f1}

@torch.no_grad()
def get_embeddings(m,ldr):
    # ознаки з fc1
    m.eval(); FE=[]
    for x,_ in ldr:
        x=x.to(DEVICE)
        _,f=m(x,return_feat=True)
        FE.append(f.detach().cpu().numpy())
    return np.concatenate(FE,0)

def fit_on_indices(indices):
    # тренуємо модель на підмножині train_full
    m=create_model()
    ldr=DataLoader(Subset(train_full,list(indices)),batch_size=BATCH_SIZE,shuffle=True,num_workers=2)
    opt=torch.optim.AdamW(m.parameters(),lr=LR)
    t0=time.time()
    for _ in range(EPOCHS):
        train_epoch(m,ldr,opt)
    elapsed=time.time()-t0
    acc,_,mtr=evaluate(m,DataLoader(test_ds,batch_size=BATCH_SIZE,shuffle=False,num_workers=2))
    return m,acc,elapsed,mtr
```

```python
# тренуємо на всьому train
ive_model, ive_acc, ive_time, ive_more = fit_on_indices(train_idx)
ive_samples = len(train_idx)
ive_cost = ive_samples * ANNO_COST_PER_SAMPLE

# результати
print({
    "ive_samples": ive_samples,
    "ive_accuracy": round(ive_acc,4),
    "ive_time_min": round(ive_time/60,2),
    "ive_cost_$": round(ive_cost,2)
})
```

```python
def random_sampling(pool_ids, batch_size):
    # випадковий вибір з пулу
    k=min(batch_size,len(pool_ids))
    return np.random.choice(pool_ids,size=k,replace=False)

def uncertainty_sampling(model, ds_pool, batch_size):
    # ентропія прогнозів
    ld=DataLoader(ds_pool,batch_size=BATCH_SIZE,shuffle=False,num_workers=2)
    _,P,_=evaluate(model,ld)
    ent=shannon_entropy((P+1e-12).T)
    order=np.argsort(-ent)
    sel=order[:min(batch_size,len(order))]
    return np.array(ds_pool.indices)[sel]

def query_by_committee(models, ds_pool, batch_size):
    # vote entropy
    ld=DataLoader(ds_pool,batch_size=BATCH_SIZE,shuffle=False,num_workers=2)
    votes=[evaluate(m,ld)[1].argmax(1) for m in models]
    votes=np.stack(votes,0)
    n_models=votes.shape[0]
    vmat=np.zeros((votes.shape[1],10),dtype=np.int32)
    for i in range(n_models):
        v=votes[i]; vmat[np.arange(len(v)),v]+=1
    pv=vmat/n_models
    ve=shannon_entropy((pv+1e-12).T)
    order=np.argsort(-ve)
    sel=order[:min(batch_size,len(order))]
    return np.array(ds_pool.indices)[sel]

def diversity_sampling(model, ds_pool, batch_size):
    # k-means на ембеддінгах
    ld=DataLoader(ds_pool,batch_size=BATCH_SIZE,shuffle=False,num_workers=2)
    FE=get_embeddings(model,ld)
    km=KMeans(n_clusters=min(batch_size,max(2,batch_size)),n_init='auto',random_state=SEED).fit(FE)
    labs=km.labels_; sel=[]
    for c in np.unique(labs):
        idx=np.where(labs==c)[0]
        center=km.cluster_centers_[c][None,:]
        d=cosine_distances(FE[idx],center).ravel()
        sel.append(idx[np.argmin(d)])
    if len(sel)<batch_size:
        rest=list(set(range(len(FE)))-set(sel))
        add=np.random.choice(rest,size=min(batch_size-len(sel),len(rest)),replace=False)
        sel=list(sel)+list(add)
    return np.array(ds_pool.indices)[sel[:batch_size]]

def hybrid_sampling(model, ds_pool, batch_size, alpha=0.7):
    # поєднуємо ентропію та "відстань до центрів"
    ld=DataLoader(ds_pool,batch_size=BATCH_SIZE,shuffle=False,num_workers=2)
    _,P,_=evaluate(model,ld)
    ent=shannon_entropy((P+1e-12).T)
    ent=(ent - np.min(ent)) / (np.ptp(ent) + 1e-9)

    FE=get_embeddings(model,ld)
    km=KMeans(n_clusters=min(batch_size,max(2,batch_size)),n_init='auto',random_state=SEED).fit(FE)
    D=cosine_distances(FE,km.cluster_centers_).min(axis=1)
    D=(D - np.min(D)) / (np.ptp(D) + 1e-9)
    score=alpha*ent+(1-alpha)*D
    order=np.argsort(-score)
    sel=order[:min(batch_size,len(order))]
    return np.array(ds_pool.indices)[sel]
```

```python
def calculate_reduction_rate(n_active, n_ive, target_acc):
    # RR = (N_ive - N_active) / N_ive * 100%
    if n_ive==0: return 0.0
    return float(max(0.0,(n_ive-n_active)/n_ive*100.0))

def samples_to_threshold(accuracies, samples, threshold=0.90):
    # мін. к-сть зразків для порогу
    for s,a in zip(samples,accuracies):
        if a>=threshold:
            return int(s)
    return None

def cost_to_performance(total_cost, accuracy):
    # нижче краще
    if accuracy<=1e-9: return float("inf")
    return float(total_cost/accuracy)
```

```python
def calculate_information_gain(H_before, H_after):
    # IG = H(Y) - H(Y|X_selected)
    return float(max(0.0, H_before - H_after))

def calculate_diversity_score(embeddings):
    # середня попарна косинусна відстань
    if embeddings.shape[0]<2: return 0.0
    D=cosine_distances(embeddings)
    iu=np.triu_indices_from(D,k=1)
    return float(D[iu].mean())

def calculate_pool_coverage(selected_embeddings, pool_embeddings, n_clusters=10):
    # частка кластерів пулу, що покриті вибіркою
    if len(pool_embeddings)==0 or len(selected_embeddings)==0: return 0.0
    km=KMeans(n_clusters=min(n_clusters,len(pool_embeddings)),n_init='auto',random_state=SEED)
    pool_labels=km.fit_predict(pool_embeddings)
    sel_labels=km.predict(selected_embeddings)
    return float(len(np.unique(sel_labels))/len(np.unique(pool_labels)))
```

```python
def run_strategy(name):
    # ініціалізуємо розмічене 5%
    ids=np.array(train_idx); np.random.shuffle(ids)
    labeled=set(ids[:INIT_LABELED].tolist())
    pool=list(set(train_idx)-labeled)

    # стартова модель
    model,acc,t,mm = fit_on_indices(list(labeled))
    total_time=t
    total_cost=len(labeled)*ANNO_COST_PER_SAMPLE

    logs=[]
    for step in range(1, AL_STEPS+1):
        ds_pool=Subset(train_full, list(pool))

        # вибірка за стратегією
        if name=='random':
            pick=random_sampling(np.array(pool),BATCH_AL).tolist()
        elif name=='uncertainty':
            pick=uncertainty_sampling(model,ds_pool,BATCH_AL).tolist()
        elif name=='qbc':
            m2=create_model(SEED+1); m3=create_model(SEED+2)
            ldl=DataLoader(Subset(train_full,list(labeled)),batch_size=BATCH_SIZE,shuffle=True,num_workers=2)
            o2=torch.optim.AdamW(m2.parameters(),lr=LR); o3=torch.optim.AdamW(m3.parameters(),lr=LR)
            for _ in range(max(1,EPOCHS-1)):
                train_epoch(m2,ldl,o2); train_epoch(m3,ldl,o3)
            pick=query_by_committee([model,m2,m3],ds_pool,BATCH_AL).tolist()
        elif name=='diversity':
            pick=diversity_sampling(model,ds_pool,BATCH_AL).tolist()
        elif name=='hybrid':
            pick=hybrid_sampling(model,ds_pool,BATCH_AL,alpha=0.7).tolist()

        # до навчання: IG частина 1 (ентропія по тесту)
        acc_b,P_b,_=evaluate(model,DataLoader(test_ds,batch_size=BATCH_SIZE,shuffle=False,num_workers=2))
        H_b=shannon_entropy((P_b+1e-12).T).mean()

        # метрики різноманіття
        FE_sel=get_embeddings(model,DataLoader(Subset(train_full,pick),batch_size=BATCH_SIZE,shuffle=False,num_workers=2))
        FE_pool=get_embeddings(model,DataLoader(Subset(train_full,pool),batch_size=BATCH_SIZE,shuffle=False,num_workers=2)) if len(pool)>0 else np.zeros((0,256))
        div=calculate_diversity_score(FE_sel)
        cov=calculate_pool_coverage(FE_sel,FE_pool,10) if len(FE_pool)>0 else 0.0

        # додаємо розмітку та перенавчаємо
        labeled |= set(pick)
        pool = list(set(pool)-set(pick))
        model,acc,t,mm = fit_on_indices(list(labeled))
        total_time += t
        total_cost += len(pick)*ANNO_COST_PER_SAMPLE

        # IG частина 2
        _,P_a,_=evaluate(model,DataLoader(test_ds,batch_size=BATCH_SIZE,shuffle=False,num_workers=2))
        H_a=shannon_entropy((P_a+1e-12).T).mean()
        IG=calculate_information_gain(H_b,H_a)

        # лог поточної ітерації
        logs.append({
            "step":step,
            "labeled":len(labeled),
            "acc":float(acc),
            "precision":float(mm["precision"]),
            "recall":float(mm["recall"]),
            "f1":float(mm["f1"]),
            "time_min":total_time/60.0,
            "cost":total_cost,
            "info_gain":IG,
            "diversity":div,
            "coverage":cov
        })

    df=pd.DataFrame(logs)

    # зведені показники для стратегії
    s_thr = samples_to_threshold(df["acc"].values, df["labeled"].values, threshold=TARGET_ACC)
    n_active = s_thr if s_thr is not None else df["labeled"].iloc[-1]
    rr = calculate_reduction_rate(n_active, len(train_idx), TARGET_ACC)
    cpr = cost_to_performance(df["cost"].iloc[-1], df["acc"].iloc[-1])
    return df, {"samples_to_90": s_thr, "reduction_rate_%": rr, "cost_to_performance": cpr}

# запускаємо 5 стратегій
results = {}
rows = []
for name in ["random","uncertainty","qbc","diversity","hybrid"]:
    df, s = run_strategy(name)
    results[name]=df
    rows.append({
        "method":name,
        **s,
        "final_acc":df["acc"].iloc[-1],
        "final_f1":df["f1"].iloc[-1],
        "total_time_min":df["time_min"].iloc[-1],
        "total_cost":df["cost"].iloc[-1],
        "avg_info_gain":df["info_gain"].mean(),
        "avg_diversity":df["diversity"].mean(),
        "avg_coverage":df["coverage"].mean()
    })

summary_df = pd.DataFrame(rows)
summary_df
```

```python
def calculate_total_efficiency(accuracy, cost, time_min,
                               alpha=0.5, beta=0.3, gamma=0.2,
                               cost_min=None, cost_max=None,
                               time_min_v=None, time_max_v=None):
    # нормалізація cost/time та обчислення
    cmin = cost if cost_min is None else cost_min
    cmax = cost if cost_max is None else cost_max
    tmin = time_min if time_min_v is None else time_min_v
    tmax = time_min if time_max_v is None else time_max_v
    c_norm = 0.0 if cmax==cmin else (cost-cmin)/(cmax-cmin)
    t_norm = 0.0 if tmax==tmin else (time_min-tmin)/(tmax-tmin)
    inv_c = 1.0 - c_norm
    inv_t = 1.0 - t_norm
    return float(alpha*accuracy + beta*inv_c + gamma*inv_t)

# обчислюємо ефективність для 3 сценаріїв (по рядках summary_df)
cmin,cmax = summary_df["total_cost"].min(), summary_df["total_cost"].max()
tmin,tmax = summary_df["total_time_min"].min(), summary_df["total_time_min"].max()

summary_df = summary_df.copy()
summary_df["eff_research"] = summary_df.apply(lambda r: calculate_total_efficiency(
    r["final_acc"], r["total_cost"], r["total_time_min"], 0.7,0.2,0.1, cmin,cmax,tmin,tmax), axis=1)
summary_df["eff_startup"]  = summary_df.apply(lambda r: calculate_total_efficiency(
    r["final_acc"], r["total_cost"], r["total_time_min"], 0.4,0.4,0.2, cmin,cmax,tmin,tmax), axis=1)
summary_df["eff_prod"]     = summary_df.apply(lambda r: calculate_total_efficiency(
    r["final_acc"], r["total_cost"], r["total_time_min"], 0.5,0.2,0.3, cmin,cmax,tmin,tmax), axis=1)

summary_df
```
